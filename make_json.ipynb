```python
import json
dataset = {'categories': [], 'annotations': [], 'images': []}

path = './train/train/'
```

```python
# create list length = 65
with open('custom.txt') as f:
  content = f.read()

for class_i, class_name in enumerate(content.split(sep='\n')):
  if class_name != '':
    dataset['categories'].append({'id':class_i,'name':class_name,'supercategory':'RoadSign'})
```

```python
dataset
```

```python
# txt file load
import glob
train_files = glob.glob(path+'*.txt')
```

```python
for train_file in train_files:
  if train_file[14:] != 'filename.txt':
    dataset['images'].append({'filename':train_file[14:20]+'.jpg', 'id':int(train_file[14:20]), 'width':1920, 'height':1200})
```

```python
dataset
```

```python
import time
start = time.time()

global count
count = 0
width = 1920
height = 1200
for train_file in train_files:
  if train_file[14:] != 'filename.txt':
    with open(path + train_file[14:]) as f:
      annos = f.readlines()
      image_id = int(train_file[14:20])
      for i, anno in enumerate(annos):
        parts = anno.strip().split()
        cls_id = int(parts[0])
        x_min = float(parts[1]) * width
        x_max = (float(parts[1]) + float(parts[3])) * width 
        y_min = float(parts[2]) * height
        y_max = (float(parts[2]) + float(parts[4])) * height
        width_b = max(0, x_max - x_min)
        height_b = max(0, y_max - y_min)
        dataset['annotations'].append({
            'area':width * height,
            'bbox':[x_min, y_min, width_b, height_b],
            'category_id':cls_id,
            'id':count,
            'image_id':image_id,
            'iscrowd':0,
            'segmentation':[[x_min, y_min, x_max, y_min, x_max, y_max, x_min, y_max]]
        })
        count += 1

print('걸린시간:{:.3f}'.format(time.time() - start))
```

```python
json_name = './instances_train.json'
with open(json_name, 'w') as f:
    json.dump(dataset, f)
```
